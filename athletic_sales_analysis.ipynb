### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
import pandas as pd

# Read the CSV files into DataFrames
df_2020 = pd.read_csv(r'Z:\Columbia\athletic_sales_analysis\Resources\athletic_sales_2020.csv')
df_2021 = pd.read_csv(r'Z:\Columbia\athletic_sales_analysis\Resources\athletic_sales_2021.csv')

```

```python
# Display the 2020 sales DataFrame
# Display the 2020 sales DataFrame with specified categories and limited rows
df_2020_display = df_2020[['retailer', 'retailer_id', 'invoice_date', 'region', 'state', 'city', 
                           'product', 'price_per_unit', 'units_sold', 'total_sales', 
                           'operating_profit', 'sales_method']]

# Display the first 5 rows and the last 5 rows
display(df_2020_display.head(5))  # First 5 rows
display(df_2020_display.tail(5))  # Last 5 rows
```

```python
# Display the 2021 sales DataFrame
# Display the 2021 sales DataFrame with specified categories and limited rows
df_2021_display = df_2021[['retailer', 'retailer_id', 'invoice_date', 'region', 'state', 'city', 
                           'product', 'price_per_unit', 'units_sold', 'total_sales', 
                           'operating_profit', 'sales_method']]

# Display the first 5 rows and the last 5 rows
display(df_2021_display.head(5))  # First 5 rows
display(df_2021_display.tail(5))  # Last 5 rows
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
print(df_2020.info())
```

```python
# Check the 2021 sales data types.
print(df_2021.info())
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales = pd.concat([df_2020, df_2021], axis=0).reset_index(drop=True)

# Display the first 5 rows and the last 5 rows
display(combined_sales.head(5))  # First 5 rows
display(combined_sales.tail(5))  # Last 5 rows
```

```python
# Check if any values are null.
null_values = combined_sales.isnull().sum()

# Display the result
print(null_values)
```

```python
# Check the data type of each column
print(combined_sales.info())
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales['invoice_date'] = pd.to_datetime(combined_sales['invoice_date'], errors='coerce')

# Display the first few rows of the updated DataFrame
display(combined_sales.head())
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_sales.info())
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Group by 'region', 'state', and 'city', summing the 'units_sold' column
grouped_sales = combined_sales.groupby(['region', 'state', 'city'])['units_sold'].sum()

# Rename the resulting sum column to "Total_Products_Sold"
grouped_sales = grouped_sales.rename("Total_Products_Sold")

# Sort the results to show the top 5 cities with the most products sold
top_5_sales = grouped_sales.sort_values(ascending=False).head(5)

# Convert to DataFrame to display in table format
top_5_sales_df = top_5_sales.reset_index()

# Display the result in a proper table format
import pandas as pd
pd.set_option('display.colheader_justify', 'center')  # Optional for better table formatting

# Display the top 5 results in table format
display(top_5_sales_df)
```

#### Using `pivot_table`

```python
# Create a pivot table to show the total products sold for region, state, and city
pivot_table = pd.pivot_table(combined_sales, 
                             values='units_sold', 
                             index=['region', 'state', 'city'], 
                             aggfunc='sum')

# Rename the column to "Total_Products_Sold"
pivot_table = pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'})

# Sort the results to show the top 5
top_5_pivot = pivot_table.sort_values(by='Total_Products_Sold', ascending=False).head(5)

# Convert to DataFrame to display in table format
top_5_pivot_df = top_5_pivot.reset_index()

# Display the result in a proper table format
import pandas as pd
pd.set_option('display.colheader_justify', 'center')  # Optional for better table formatting

# Display the top 5 results in table format
display(top_5_pivot_df)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

#### Using `groupby`

```python
# Group by 'retailer', 'region', 'state', and 'city', summing the 'units_sold' for women's athletic footwear
grouped_womens_footwear_sales = womens_athletic_footwear.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum()

# Rename the resulting sum column to "Womens_Footwear_Units_Sold"
grouped_womens_footwear_sales = grouped_womens_footwear_sales.rename("Womens_Footwear_Units_Sold")

# Sort the results to show the top 5 retailers with the most units sold for women's athletic footwear
top_5_womens_footwear_sales = grouped_womens_footwear_sales.sort_values(ascending=False).head(5)

# Convert the result into a DataFrame for proper table formatting
top_5_womens_footwear_sales_df = top_5_womens_footwear_sales.reset_index()

# Display the result as a table
import pandas as pd
pd.set_option('display.colheader_justify', 'center')  # Optional for better table formatting

# Display the result in table format
display(top_5_womens_footwear_sales_df)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_womens_footwear_sales = pd.pivot_table(womens_athletic_footwear, 
                                             values='units_sold', 
                                             index=['retailer', 'region', 'state', 'city'], 
                                             aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_womens_footwear_sales = pivot_womens_footwear_sales.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results
top_5_pivot_womens_footwear_sales = pivot_womens_footwear_sales.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)

# Display the result in a proper table format
import pandas as pd
pd.set_option('display.colheader_justify', 'center')  # Optional for better table formatting

# Display the top 5 results in a proper table format
display(top_5_pivot_womens_footwear_sales)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column as the index, and the "total_sales" as the values.
pivot_table_sales_by_date = pd.pivot_table(womens_athletic_footwear, 
                                           values='total_sales', 
                                           index='invoice_date', 
                                           aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_sales_by_date = pivot_table_sales_by_date.rename(columns={'total_sales': 'Total Sales'})

# Show the table and sort by "Total Sales" in descending order
top_sales_days = pivot_table_sales_by_date.sort_values(by='Total Sales', ascending=False).head(10)

# Display the result in a proper table format
import pandas as pd
pd.set_option('display.colheader_justify', 'center')  # Optional for better table formatting

# Display the top 10 days with the most sales in table format
display(top_sales_days)

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
resampled_sales_by_day = pivot_table_sales_by_date.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
sorted_sales_by_day = resampled_sales_by_day.sort_values(by='Total Sales', ascending=False)

# Display the result in a proper table format
import pandas as pd
pd.set_option('display.colheader_justify', 'center')  # Optional for better table formatting

# Display top 10 days with the most sales
display(sorted_sales_by_day.head(10))
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
resampled_sales_by_week = pivot_table_sales_by_date.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
sorted_sales_by_week = resampled_sales_by_week.sort_values(by='Total Sales', ascending=False)

# Display the result in a proper table format
import pandas as pd
pd.set_option('display.colheader_justify', 'center')  # Optional for better table formatting

# Display top 10 weeks with the most sales
display(sorted_sales_by_week.head(10))
```
